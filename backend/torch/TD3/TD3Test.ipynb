# Training

```python
import itertools
import datetime
import sys
sys.path.insert(0,'../../envs/')
sys.path.insert(0,'../core/')

import os
from utils import *
from PegRobot2D import Frontend, WINDOW_X, WINDOW_Y
import numpy as np
import torch
from td3 import TD3Agent
from tensorboardX import SummaryWriter
import matplotlib.pyplot as plt
from rl_batch_trainer import BatchRLAlgorithm

save_dir = "models/"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
actor_path = save_dir
if not os.path.exists(actor_path):
    os.makedirs(actor_path)
critic_path = save_dir
if not os.path.exists(critic_path):
    os.makedirs(critic_path)

variant = dict(
        algorithm="TD3",
        version="normal",
        seed = 0,
        save_model = True,

        algorithm_kwargs=dict(
            num_epochs= 50,
            num_eval_steps_per_epoch= 1000,
            num_train_loops_per_epoch = 5,
            num_trains_per_train_loop = 250, 
            num_expl_steps_per_train_loop = 2500,
            min_num_steps_before_training = 10000, # Random exploration steps Initially
            max_path_length= 250,
            batch_size = 128,
            prioritised_experience = True,
            importance_sampling = True,
            until_done = True,
        ),

        trainer_kwargs=dict(
            gamma=0.99,
            tau=0.005,
            noise = "normal",
            exploration_noise = 0.1, #0.1,
            hidden_size = 128,       
            delayed_policy_steps = 2,
            replay_buffer_size = int(3e5), 
            noise_std = 0.2, #0.2,
            noise_bound = 0.5,
            critic_lr = 3e-4,
            actor_lr = 3e-4,
        ),

        env_args = [
            WINDOW_X,
            WINDOW_Y,
            "Peg 2D Robot"
            ],

        env_kwargs = dict(
            vsync = False,
            resizable = False,
            visible = False
            )
)

# Environment
env = Frontend(*variant['env_args'], **variant['env_kwargs'])
if variant['algorithm'] is "SAC" or "TD3":
    env.denorm_process = False # No need to denorm because in SAC the gaussian policies are already scaled up
elif variant['algorithm'] is "DDPG":
    env.denorm_process = True # Action outputs are in [-1, 1] so we need to denormalise them to pass into the envirnoment

torch.manual_seed(variant['seed'])
np.random.seed(variant['seed'])

# Agent
num_actions = env.num_actions
num_inputs = env.num_states
action_range = env.action_range

agent = TD3Agent(num_inputs, num_actions, action_range, **variant['trainer_kwargs'])

# Tensorboard
log_dir = 'runs/{}_{}_{}'.format(datetime.datetime.now().strftime("%Y-%m-%d_%H-%M-%S"),
                                        "Peg2DRobot", variant['algorithm'], variant['trainer_kwargs']['noise'])
writer = SummaryWriter(logdir=log_dir)

# training
RL_trainer = BatchRLAlgorithm(**variant['algorithm_kwargs']) 
RL_trainer.train(env, agent, writer)

# Save model 
if variant['save_model']:
    agent.save_model("Peg2D")
```

```python
import itertools
import datetime
import sys
sys.path.insert(0,'../../envs/')
sys.path.insert(0,'../core/')

import os
from utils import *
from PegRobot2D import Frontend, WINDOW_X, WINDOW_Y
import numpy as np
import torch
from td3 import TD3Agent
from tensorboardX import SummaryWriter
import matplotlib.pyplot as plt
from rl_batch_trainer import BatchRLAlgorithm

save_dir = "models/"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
actor_path = save_dir
if not os.path.exists(actor_path):
    os.makedirs(actor_path)
critic_path = save_dir
if not os.path.exists(critic_path):
    os.makedirs(critic_path)

variant = dict(
        algorithm="TD3",
        version="normal",
        seed = 1,
        save_model = True,

        algorithm_kwargs=dict(
            num_epochs= 2,
            num_eval_steps_per_epoch= 1000,
            num_train_loops_per_epoch = 10,
            num_trains_per_train_loop= 250, # Was 100
            num_expl_steps_per_train_loop = 1000,
            min_num_steps_before_training = 5000, # Random exploration steps Initially
            max_path_length= 250,
            batch_size = 128,
            prioritised_experience = False,
            importance_sampling = False,
        ),

        trainer_kwargs=dict(
            gamma=0.99,
            tau=0.005,
            noise = "normal",
            hidden_size = 128,       
            delayed_policy_steps = 2,
            replay_buffer_size = int(2e5), # Increase this..
            noise_std = 0.1,
            noise_bound = 0.5,
            critic_lr = 1e-3,
            actor_lr = 1e-3,
        ),

        env_args = [
            WINDOW_X,
            WINDOW_Y,
            "Peg 2D Robot"
            ],

        env_kwargs = dict(
            vsync = False,
            resizable = False,
            visible = False
            )
)

# Environment
env = Frontend(*variant['env_args'], **variant['env_kwargs'])
if variant['algorithm'] is "SAC":
    env.denorm_process = False # No need to denorm because in SAC the gaussian policies are already scaled up
    
torch.manual_seed(variant['seed'])
np.random.seed(variant['seed'])

def run_policy(agent, env = None, framework = "SAC"):
    if isinstance(env, Frontend):
        del(env)
    env = Frontend(WINDOW_X, WINDOW_Y, "RoboPeg2D Simulation", vsync = False, resizable = False, visible = True)
    env.agent = agent
    if framework is "SAC" or "TD3":
        env.denorm_process = False # Necessary for SAC
    else:
        env.denorm_process = True
    env.run_policy(agent)

def load_model(self, actor_path, critic_1_path, critic_2_path):
    print('Loading models from {} and {} and {}'.format(actor_path, critic_1_path, critic_2_path))
    if actor_path is not None:
        self.actor.load_state_dict(torch.load(actor_path))
    if critic_1_path is not None and critic_2_path is not None:
        self.critic1.load_state_dict(torch.load(critic_1_path))
        self.critic2.load_state_dict(torch.load(critic_2_path))
        
env = Frontend(WINDOW_X, WINDOW_Y, "RoboPeg2D Simulation", vsync = False, resizable = False, visible = False)
    
# Agent
num_actions = env.num_actions
num_inputs = env.num_states
action_range = env.action_range

tst_agent = TD3Agent(num_inputs, num_actions, action_range, **variant['trainer_kwargs'])

load_model(tst_agent, "models/actor_Peg2D_", "models/critic_1_Peg2D_", "models/critic_2_Peg2D_")

run_policy(tst_agent, env, "TD3")
```
